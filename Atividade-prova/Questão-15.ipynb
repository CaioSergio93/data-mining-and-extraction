```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
```

```python
n_samples = 1000
n_features = 2
n_clusters = 3
```

```python
X, _ = make_blobs(n_samples=n_samples, centers=n_clusters, cluster_std=0.60, random_state=0)
```

```python
eps = 0.5  
min_samples = 5 
```

```python
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
labels = dbscan.fit_predict(X)
```

```python
if len(set(labels)) > 1:  # Verifica se há mais de um cluster
    silhouette_avg = silhouette_score(X, labels)
    print(f'Coeficiente de Silhueta: {silhouette_avg:.2f}')
else:
    print('Apenas um cluster encontrado. O coeficiente de Silhueta não é aplicável.')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', alpha=0.5, marker='o', edgecolor='k')
plt.title('Clusterização com DBSCAN')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
